### Working with NaN values
* The empty values in the file are replced by default with NaN values.

```python
# working with missing values
m = pd.read_csv('IMDB-Movie-Data.csv')
```

```python
m.head(3)
```

```python
m.isnull().sum()
```

```python
m.isnull()
```

```python
# get rid of NaN values
m.dropna().isnull().sum()     #deletes rows with null values
```

```python
m.dropna(axis=1).isnull().sum()  #deletes columns with NaN values
```

```python
m.dropna(how="any")   #deletes all rows with NaN
```

```python
m.dropna(how="all")  # drops row ONLY containing NaN as values
```

## Replacing null/NaN values

```python
#fillna()

#creating dataframe
import pandas as pd
import numpy as np

df = pd.DataFrame([[np.nan,np.nan,3,0],[np.nan,1,np.nan,5],[np.nan,3,0,np.nan],[9,np.nan,8,np.nan]],columns=['A','B','C','D'])
print(df)
```

```python
df.fillna(0)
```

```python
x={'A':1,'B':2,'C':3,'D':4}   #filling different column NaN diffently
df.fillna(value=x)
```

```python
#inplace parameter present
x={'A':10,'B':20,'C':30,'D':40}
df.fillna(value=x,inplace=True)
print(df)
```

```python
#limit parameter : if there is a gap with more than this number of consecutive NaNs, 
# it will only be partially filled. 
# If method is not specified, this is the maximum number of entries along the entire
# axis where NaNs will be filled. 
# Must be greater than 0 if not None.

df = pd.DataFrame([[np.nan,np.nan,3,0],[np.nan,1,np.nan,5],[np.nan,3,0,np.nan],[9,np.nan,8,np.nan]],columns=['A','B','C','D'])
print(df)
x={'A':10,'B':20,'C':30,'D':40}
df.fillna(10,limit=None,axis=0) #no consecutive NaN values
```

```python
df.fillna(10,limit=1)
```

### Filling mean in NaN values

```python
x={'A':10,'B':20,'C':30,'D':40}
df = pd.DataFrame([[np.nan,3,0,np.nan],[np.nan,1,np.nan,5],[np.nan,3,0,np.nan],[9,np.nan,8,np.nan]],columns=['A','B','C','D'])
print(df)
```

```python
df.fillna(df['A'].mean())   #replacing all NaN(s) with mean of col "A"
```

```python
df
#as inplace=False
```

```python
mean = df['A'].mean()
df['A'].fillna(mean,inplace=True) #fill only A with A's mean value
print(df)
```
