### Objectives of Simple Linear regression
* Model relationship b/w two variables
* Predict output for new instances<br>
_For this we will be using the __Least Square Method__._

```python
# Linear regression (Supervised learning)

import numpy as np
X=np.array([[-2],[1],[3]])  # Input data
Y=np.array([[-1],[1],[2]])  # Output varaible data
print("Input data : \n",X)
print("Output data : \n",Y)
print()
print(np.size(X))
print()

from sklearn.linear_model import LinearRegression
model = LinearRegression()  #defining model
model.fit(X,Y)   #fitting data into model

#beta1
print("Beta1 / y-Intecept = ",model.intercept_) 

#beta2
print("Beta2 / Slope = ",model.coef_)  
v=model.predict([[-2]]) #new input1
print("Prediction for -2 : ",v)
print()
v1=model.predict([[2]])  # new input2
print("Prediction for 2 : ",v1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data = pd.read_csv('houses_price.csv')
print(data)
print()

X=data[['No of Bed Rooms']]
Y=data[['Price']]

plt.scatter(X,Y)
plt.xlabel('No of Bed Rooms')
plt.ylabel('Price')
plt.show()

model = LinearRegression()
x_train,x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
print("Input training data :\n",x_train)
print("Output training data :\n",y_train)
model.fit(x_train, y_train)
pred= model.predict(x_test)
print("Beta2 / Slope = ",model.coef_) 
print("Beta1 / y-Intecept = ",model.intercept_)
v1=model.predict([[2]])  # new input 
print("Prediction for input = 2 : \n",v1)
```
